```python
import matplotlib.pyplot as plt
import numpy as np


figura, ejes = plt.subplots(2,1)
t = np.linspace(0,10,100)
ejes[0].set_ylabel('Esfuerzos')
ejes[0].plot([-3,-2,5,4,3.2,2.2,1.7,1.5,1.2,1.1,1])
ejes[1].set_ylabel('Desplazamientos')
ejes[1].plot(t,np.sin(t),'r--')
```
